```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize': (15, 10)}, font_scale=1.4)
```

```python
sex = [np.random.choice(['male', 'female']) for x in range(1000)]
```

   -   ,   

```python
math_score = [np.random.randint(32, 101) for x in range(1000)]
biology_score = [np.random.randint(85, 101) for x in range(1000)]
chemistry_score = [np.random.randint(71, 101) for x in range(1000)]
```

    ,  *  *
*    ,          c  *

```python
math_score.sort()
```

```python
data = {'sex': sex, 'math_score': math_score, 'biology_score': biology_score,
        'chemistry_score': chemistry_score}
```

     *pandas.DataFrame*

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

   DataFrame    

```python
df['total_score'] = df.math_score + df.biology_score + df.chemistry_score
```

```python
df.head()
```

```python
ax = sns.lineplot(x='total_score',
                  y='math_score',
                  hue="sex",
                  data=df)

ax.set(xlabel='Total Score', ylabel='Math Score')
```

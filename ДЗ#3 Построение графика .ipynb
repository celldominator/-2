Построим график, похожий на этот 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize': (15, 10)}, font_scale=1.4)
```

Сгенерируем свой набор данных:

 Создадим группирующую переменную - пол

```python
sex = [np.random.choice(['male', 'female']) for x in range(1000)]
```

Создадим количественные переменные - баллы по математике, биологии и химии

```python
math_score = [np.random.randint(32, 101) for x in range(1000)]
biology_score = [np.random.randint(85, 101) for x in range(1000)]
chemistry_score = [np.random.randint(71, 101) for x in range(1000)]
```

Чтобы график не выглядел хаотично, отсортируем *баллы по математике*ю
*Мы сортируем именно эти баллы, потому что будет строится график связи баллов по математике c итоговыми баллами*

```python
math_score.sort()
```

Создадим словарь содержащий название и набор данных для каждой нашей группы

```python
data = {'sex': sex, 'math_score': math_score, 'biology_score': biology_score,
        'chemistry_score': chemistry_score}
```

На основе этого словаря создадим *pandas.DataFrame*

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

Добавим в итоговый DataFrame колонку с итоговыми баллами

```python
df['total_score'] = df.math_score + df.biology_score + df.chemistry_score
```

```python
df.head()
```

Построим график зависимости итогового балла от баллов по математике

```python
ax = sns.lineplot(x='total_score',
                  y='math_score',
                  hue="sex",
                  data=df)

ax.set(xlabel='Total Score', ylabel='Math Score')
```
